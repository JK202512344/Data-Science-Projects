Calculate the median rating grouped by category

```python
df_category_median = df.groupby('Category')['Review Rating'].median()
df_category_median
```

Identify the records with Review Rating as null

```python
df[df['Review Rating'].isnull()]
```

code fills the missing values in the Review Rating column by replacing each null value with the median rating of its respective category.

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df['Review Rating'].isnull().sum()
```

```python
df.isnull().sum()
```

converts all column names to lowercase, replaces spaces with underscores, and rename the column purchase_amount_(usd) to purchase_amount for consistency

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df= df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})

df.columns
```

create an age group column by dividing the age column into four quantile-based bins and labeling them as Young Adult, Adult, Middle Aged, and Senior.

```python
labels = {'Young Adult','Adult','Middle Aged','Senior'}
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df.head()
```

Identify unique values in frequency_of_purchases column

```python
df['frequency_of_purchases'].unique()
```

map each purchase frequency category to its equivalent number of days and creates a new column, purchase_frequency_days, by converting the values in frequency_of_purchases using the defined frequency_mapping dictionary.

```python
frequency_mapping = {
    'Fortnightly':14, 
    'Weekly':7, 
    'Annually':365, 
    'Quarterly':90 , 
    'Bi-Weekly':14,
    'Monthly':30, 
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df.frequency_of_purchases.map(frequency_mapping)
```

```python
from sqlalchemy import create_engine, text

USER = "root"              # your MySQL username
PASSWORD = "root" # your MySQL password
HOST = "localhost"         # or IP / hostname
PORT = 3306                # default MySQL port
DB_NAME = "customer_behavior"      # database name (will create if not exists)
TABLE_NAME = "customers_data"   # new table name to create

# ---------- 3. CREATE ENGINE (SERVER-LEVEL, NO DB YET) ----------
# Connect to MySQL server first (no specific DB)
server_url = f"mysql+mysqlconnector://{USER}:{PASSWORD}@{HOST}:{PORT}/"
server_engine = create_engine(server_url)

# ---------- 4. CREATE DATABASE IF NOT EXISTS ----------
with server_engine.connect() as conn:
    conn.execute(text(f"CREATE DATABASE IF NOT EXISTS {DB_NAME}"))
    conn.commit()

# ---------- 5. CONNECT TO THE TARGET DATABASE ----------
db_url = f"mysql+mysqlconnector://{USER}:{PASSWORD}@{HOST}:{PORT}/{DB_NAME}"
engine = create_engine(db_url)

# ---------- 6. WRITE DATAFRAME TO NEW TABLE ----------
# if_exists options:
#   'fail'  -> error if table exists
#   'replace' -> drop & recreate table
#   'append'  -> insert into existing table
df.to_sql(
    TABLE_NAME,
    con=engine,
    if_exists="fail",   # change to 'replace' or 'append' if needed
    index=False         # dont write DataFrame index as a column
)

print(f"DataFrame written to {DB_NAME}.{TABLE_NAME} successfully!")
```
